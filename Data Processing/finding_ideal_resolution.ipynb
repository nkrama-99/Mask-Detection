```
width_values = []
height_values = []
mask_width_values = []
mask_height_values = []
nomask_width_values = []
nomask_height_values = []

odd_channels = []

mask_path = '/content/drive/My Drive/Colab Notebooks/Project/compute_mean_median_resolution/our_dataset/mask/'
nomask_path = '/content/drive/My Drive/Colab Notebooks/Project/compute_mean_median_resolution/our_dataset/nomask/'
```

```
width_values = []
height_values = []
mask_width_values = []
mask_height_values = []
nomask_width_values = []
nomask_height_values = []

odd_channels = []


print("checking mask images")
for file in os.listdir(mask_path):
    read_path = mask_path + file
    img = cv2.imread(read_path)

    print(file, img.shape)
    
    height_values.append(img.shape[0])
    width_values.append(img.shape[1])
    
    mask_height_values.append(img.shape[0])
    mask_width_values.append(img.shape[1])
    
    if img.shape[2] != 3:
      odd_channels.append('mask_'+file)
    
    img = []

print("checking no mask images")
for file in os.listdir(nomask_path):
    read_path = nomask_path + file
    img = cv2.imread(read_path)
    
    print(file, img.shape)

    height_values.append(img.shape[0])
    width_values.append(img.shape[1])
    
    nomask_height_values.append(img.shape[0])
    nomask_width_values.append(img.shape[1])

    if img.shape[2] != 3:
      odd_channels.append('nomask_'+file)
    img = []
```

```
def print_results(width_values, height_values):
    print()
    print("Odd cases")
    print(odd_channels)

    print()
    print("Total images:", len(width_values))
    print("Total images:", len(height_values))

    print()
    print("Width - Mean:", statistics.mean(width_values))
    print("Height - Mean:", statistics.mean(height_values))

    print()
    print("Width - Std Deviation:", statistics.stdev(width_values))
    print("Height - Std Deviation:", statistics.stdev(height_values))

    print()
    print("Width - Median:", statistics.median(width_values))
    print("Height - Median:", statistics.median(height_values))

    print()
    print("Width - Median High:", statistics.median_high(width_values))
    print("Height - Median High:", statistics.median_high(height_values))

    print()
    print("Width - Median Low:", statistics.median_low(width_values))
    print("Height - Median Low:", statistics.median_low(height_values))

    print()
    print("Width - Max:", max(width_values))
    print("Height - Max:", max(height_values))

    print()
    print("Width - Min:", min(width_values))
    print("Height - Min:", min(height_values))
```

```
# together
print_results(width_values, height_values)
```

```
# mask
print_results(mask_width_values, mask_height_values)
```

```
# nomask
print_results(nomask_width_values, nomask_height_values)
```

```
outlier_threshold = 20

def test_outlier():
    mean_h=statistics.mean(height_values)
    mean_w=statistics.mean(width_values)
    mask_not_to_be_deleted = 0
    mask_to_be_deleted = 0
    nomask_not_to_be_deleted = 0
    nomask_to_be_deleted = 0

    print("checking mask images")
    for file in os.listdir(mask_path):
        read_path = mask_path + file
        img = cv2.imread(read_path)

        if ((mean_h-outlier_threshold<img.shape[0]) and (mean_w-outlier_threshold<img.shape[1])):
          mask_not_to_be_deleted+=1
        else:
          mask_to_be_deleted+=1
        
        img = []

    print("checking no mask images")
    for file in os.listdir(nomask_path):
        read_path = nomask_path + file
        img = cv2.imread(read_path)
        
        if ((mean_h-outlier_threshold<img.shape[0]) and (mean_w-outlier_threshold<img.shape[1])):
          nomask_not_to_be_deleted+=1
        else:
          nomask_to_be_deleted+=1
      
        img = []

    print()
    print("mask_not_to_be_deleted:", mask_not_to_be_deleted)
    print("mask_to_be_deleted:", mask_to_be_deleted)
    print()
    print("nomask_not_to_be_deleted:", nomask_not_to_be_deleted)
    print("nomask_to_be_deleted:", nomask_to_be_deleted)
    print()
    print("total to NOT be deleted", mask_not_to_be_deleted+nomask_not_to_be_deleted)
    print("total to be deleted", mask_to_be_deleted+nomask_to_be_deleted)
```

```
outlier_threshold = 40
test_outlier()
```

```
# ********************DELETING THE IMAGES
outlier_threshold = 20

def delete_outlier():
    mean_h=statistics.mean(height_values)
    mean_w=statistics.mean(width_values)
    mask_not_to_be_deleted = 0
    mask_to_be_deleted = 0
    nomask_not_to_be_deleted = 0
    nomask_to_be_deleted = 0

    print("checking mask images")
    for file in os.listdir(mask_path):
        read_path = mask_path + file
        img = cv2.imread(read_path)

        if ((mean_h-outlier_threshold<img.shape[0]) and (mean_w-outlier_threshold<img.shape[1])):
          mask_not_to_be_deleted+=1
        else:
          mask_to_be_deleted+=1
          os.remove(mask_path+file)
        
        img = []

    print("checking no mask images")
    for file in os.listdir(nomask_path):
        read_path = nomask_path + file
        img = cv2.imread(read_path)
        
        if ((mean_h-outlier_threshold<img.shape[0]) and (mean_w-outlier_threshold<img.shape[1])):
          nomask_not_to_be_deleted+=1
        else:
          nomask_to_be_deleted+=1
          os.remove(nomask_path+file)
      
        img = []

    print()
    print("mask_not_to_be_deleted:", mask_not_to_be_deleted)
    print("mask_to_be_deleted:", mask_to_be_deleted)
    print()
    print("nomask_not_to_be_deleted:", nomask_not_to_be_deleted)
    print("nomask_to_be_deleted:", nomask_to_be_deleted)
    print()
    print("total to NOT be deleted", mask_not_to_be_deleted+nomask_not_to_be_deleted)
    print("total to be deleted", mask_to_be_deleted+nomask_to_be_deleted)
```

```
outlier_threshold = 20
delete_outlier()
```
